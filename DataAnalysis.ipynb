```python
def fonctiondelamort(toby):
    similarity = rpz.corrwith(rpz[toby])
    rating_toby = rpz[toby].copy()
    rating_toby.replace(0, pd.np.nan, inplace=True)
    sim_toby = similarity
    rating = theme_inscriptions
    
    rating_c = rating.loc[rating_toby[rating.themecode].isnull().values & (rating.traineeusername != toby)]
    rating_c_similarity = rating_c['traineeusername'].map(sim_toby)

    rating_d = rating_c.assign(similarity=rating_c_similarity)
    rating_d.sort_values("similarity", ascending=False, inplace=True)
    return rating_d.groupby("themecode").mean().sort_values("similarity", ascending=False).drop(columns="actioncode")
```
